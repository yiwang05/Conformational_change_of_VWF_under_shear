```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering

X = pd.read_pickle('shear_rate_90k')
#Hierarchical Agglomerative Clustering
clustering = AgglomerativeClustering().fit(X)
clustering
```

```python
clustering.labels_
labels = clustering.labels_

a=X[labels==1]
b=X[labels==0]

x1 = a.convex_area
y1 = a.major_axis_length
x2 = b.convex_area
y2 = b.major_axis_length

plt.figure(figsize=(10,5))
plt.plot(x1, y1, 'o', color='red', markersize=3)
plt.plot(x2, y2,'o',color='black', markersize=3)
plt.xlabel('convex area (pixel)')
plt.ylabel('major axis length (pixel)')
```

```python
clustering = AgglomerativeClustering().fit(a)
clustering

clustering.labels_
labels = clustering.labels_

c=a[labels==1]
d=a[labels==0]

x3 = c.convex_area
y3 = c.major_axis_length
x4 = d.convex_area
y4 = d.major_axis_length

plt.figure(figsize=(10,5))
plt.plot(x3, y3, 'o', color='red', markersize=3)
plt.plot(x4, y4, 'o',color='black', markersize=3)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('convex area (pixel)')
plt.ylabel('major axis length (pixel)')
```

```python
clustering = AgglomerativeClustering().fit(b)
clustering

clustering.labels_
labels = clustering.labels_

e=b[labels==1]
f=b[labels==0]

x5 = e.convex_area
y5 = e.major_axis_length
x6 = f.convex_area
y6 = f.major_axis_length

plt.figure(figsize=(10,5))
plt.plot(x5, y5, 'o', color='red', markersize=3)
plt.plot(x6, y6, 'o',color='black', markersize=3)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('convex area (pixel)')
plt.ylabel('major axis length (pixel)')
```

```python
plt.rcParams.update({'font.size': 15})

plt.figure(figsize=(10,5))
plt.plot(x5, y5, 'o', color='red', markersize=1)
plt.plot(x6, y6, 'o', color='darkred', markersize=1)
plt.plot(x3, y3, 'o', color='cornflowerblue', markersize=1)
plt.plot(x4, y4, 'o', color='navy', markersize=1)
plt.xscale('log')
plt.yscale('log')
plt.xlabel('convex area (pixel)')
plt.ylabel('major axis length (pixel)')

plt.savefig('90k_clustering_1.png')
```
